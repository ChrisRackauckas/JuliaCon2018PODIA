```julia

# Example from
# https://github.com/stan-dev/stancon_talks/tree/master/2017/Contributed-Talks/05_margossian/models/neutropenia

m_neut = @model begin

    @param begin

        CL ∈ RealDomain(lower=5.0,  upper=15.0, init=10.0)
        Q  ∈ RealDomain(lower=5.0,  upper=40.0, init=15.0)
        V1 ∈ RealDomain(lower=15.0, upper=70.0, init=35.0)
        V2 ∈ RealDomain(lower=30.0, upper=250.0, init=105.0)

        σ ∈ RealDomain(lower=1e-8, upper=30.0, init=1.0)

        mtt ∈ RealDomain(lower=30.0,     upper=250.0, init=125.0)
        circ0 ∈ RealDomain(lower=2.0,    upper=12.0,  init=5.0)
        α ∈ RealDomain(lower=0.0002, upper=0.02,  init=0.002)
        γ ∈ RealDomain(lower=0.08,    upper=0.4,   init=0.17)

        σN ∈ RealDomain(lower=1e-8, upper=30.0, init=1.0)

        ka ∈ RealDomain(lower=1e-8,  upper=100.0,init=1.0) # no prior defined?
    end

    

    @collate begin
        k10 = CL / V1
        k12 = Q / V1
        k21 = Q / V2

        ktr = 4 / mtt
    end

    @init begin
        Depot      = 0.0
        Central    = 0.0
        Peripheral = 0.0         
        Prol     = circ0
        Transit1 = circ0
        Transit2 = circ0
        Transit3 = circ0
        Circ     = circ0
    end
        
    @dynamics begin
        dDepot      = -ka * Depot
        dCentral    =  ka * Depot - (k10 + k12) * Central + k21 * Peripheral
        dPeripheral =  k12 * Central - k21 * Peripheral

        dProl = ktr * Prol * ((1 - α*(Depot/V1)) * ((circ0/Circ)^γ) - 1)
        dTransit1 = ktr * (Prol - Transit1)
        dTransit2 = ktr * (Transit1 - Transit2)
        dTransit3 = ktr * (Transit2 - Transit3)
        dCirc     = ktr * (Transit3 - Circ)
    end

    @post begin
        cHat = Central / V1
        neutHat = Circ
    end

    @error begin
        cHat = max(Central,0) / V1
        neutHat = Circ

        logc ~ Normal(log(cHat), σ)
        logn ~ Normal(log(neutHat), σN)
    end
end
```

```julia
x0 = PKPDSimulator.init_param(m_neut)

sol,col = PKPDSimulator.pkpd_solve(m_neut, subject, x0, ())
```
